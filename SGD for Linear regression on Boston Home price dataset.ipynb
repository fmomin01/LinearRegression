```
from sklearn.linear_model import SGDRegressor
import seaborn as sns
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

def sklearn_SGDRegressor(X_Train,Y_Train,X_Test,Y_Test,alpha,learning_rate,n_iterations):

  linear_regression=SGDRegressor(alpha=alpha,learning_rate=learning_rate,n_iter_no_change=n_iterations)
  linear_regression.fit(X_Train,Y_Train)

  Y_Predicted=linear_regression.predict(X_Test)

  plt.scatter(Y_Test,Y_Predicted)
  plt.title('scatter plot between actual y and predicted y')
  plt.xlabel('Actual Y')
  plt.ylabel('Predicted Y')
  plt.grid(b=True, linewidth=0.5)
  plt.show()

  return Y_Predicted,n_iterations,linear_regression.coef_, linear_regression.intercept_
```

```
y_predicted,n_iter,weight_sgd,intercept_sgd=sklearn_SGDRegressor(X_Train,Y_Train,X_Test,Y_Test,0.0001,'constant',1)
```

```
sgd_error=mean_squared_error(Y_Test,y_predicted)
print('Mean Square Error(MSE)=', sgd_error)
print('Number of iteration=', n_iter)
```

### Using Manual Approach to find the Linear Regression

```
def find_differential_L_wrt_W(X_Data,Y_Data,W_tranpose,K,b,d):

  # Calcuate Differential of L w.r.t W
  sum_diff = np.zeros(shape=(1,d))
  for i in range(K):
    second_term = (Y_Data.iloc[i] - b - np.dot(W_tranpose,X_Data[i]))
    final_term = X_Data[i] * second_term    
    sum_diff += final_term  
  sum_diff *= (-2/K)
  return sum_diff
```

```
def find_differential_L_wrt_b(X_Data,Y_Data,W_Transpose,K,b,d):
  #Calculate differential of L w.r.t b
  sum_diff = 0
  for i in range(K):
    final_term = Y_Data.iloc[i] - b - np.dot(W_Transpose,X_Data[i])
    sum_diff += final_term
  sum_diff *= (-2/K)
  return sum_diff
```

```
def manual_predict(X_Test,W,b):
  Y_Predicted=[]

  for i in range(len(X_Test)) :
    #predict = (np.dot(W,X_Test[i]) + b)
    y=np.asscalar(np.dot(W,X_Test[i]) + b)
    Y_Predicted.append(y)
  return np.array(Y_Predicted)
```

```
def draw_scatterPlot(Y_Test,Y_Predicted) :
  plt.scatter(Y_Test,Y_Predicted)
  plt.title('scatter plot between actual y and predicted y')
  plt.xlabel('Actual Y')
  plt.ylabel('Predicted Y')
  plt.grid(b=True, linewidth=0.5)
  plt.show()
```

```
def find_and_transform(X,Y,K,n_iterations,threshold,power_t,lr_rate) :
  n,d=X.shape
  b=0
  weight = np.random.randn(1,d)
  r,iter_ = 1,1
  n=1
  while(n<=n_iterations):
    w_prev = weight
    b_prev = b
    X_data=X.sample(K)
    X_new=X_data[X_data.columns[0:]].values
    Y_new = Y

    Diff_W = find_differential_L_wrt_W(X_new,Y_new,w_prev,K,b_prev,d)
    Diff_b = find_differential_L_wrt_b(X_new,Y_new,w_prev,K,b_prev,d)

        #Calculate Wj+1 & bj+1

    Wj_plus_1 = w_prev - (r*Diff_W)
    bj_plus_1= b_prev - (r*Diff_b)

    if(iter_ >= n_iterations):      
      break

    #r = r/2
    iter_ +=1
    n=n+1
    

  print('*************************************')
  
  print('W[j+1] :' , Wj_plus_1)
  print('b[j+1] :' , bj_plus_1)
  #print('r :' , r)
  print('*************************************')  
  return Wj_plus_1,bj_plus_1          
```

```
W_star,b_star=find_and_transform(manual_training,Y_Train,50,1,0.1,0.25,0.01)
```

```
Y_Predicted=manual_predict(X_Test, W_star,b_star)
draw_scatterPlot(Y_Test,Y_Predicted)
manual_error=mean_squared_error(Y_Test,Y_Predicted)
print('Mean Square Error(MSE) Using Manual Approach =', manual_error)
```
